<h3><b> Data Dictionary : <b></h3>

*   Area(total) : Total area of the plot
*   Trooms : Total Number of rooms in the house
*   Nbedrooms : Number of bedroom in the house
*   Nbwashrooms : Number of washroom attached with bedroom
*   washrooms : total number of washroom in the house
* Roof : Does the house has roof(yes or no)
* Roof(Area) : Total area of the terrace
* Lawn(Area) : Area of the lawn including garden and parking
* Nfloor : Number of floors in the house
* API : Air purity index api is in percentage for example if api is 85 it means its 85% of the standard api for the city.
* ANB : Amenities near by amenities like hospital, park , multiplex , malls etc within 2 miles.
* Expected price : price expected by seller.
* Grade : Grade provided by company that depends on the condition and other features mentioned in the data.*(Note A is best garde and E is worst.)*

<h2><b> Spliting the test and train data <b></h2>

```python
# Test data
x_test = test.iloc[:,:-1]  # -->  Only X taken
```

```python
# Training Data Set
x_train = train.iloc[:,:-1]  # -->  X
y_train = train.iloc[:,-1]   # -->  Y
```

```python
x_train
```

<h3><b> Scalling the Data<b></h3>

```python
# Standard scaller
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

<h3><b> Machine Learning Model <b></h3>

```python
# Using Support Vectore Machine (SVM)
from sklearn.svm import SVC
svc = SVC(C = 230,
          gamma =  0.023,
          kernel = "rbf",
          random_state = 17)
svc.fit(x_train, y_train)
n_pred = svc.predict(x_test)
```

<h2><b> This code is for to get this whole Def function into a variable called "code" <b></h2>

```python
code = '''
def validate_input(prompt, min_value, max_value):
    while True:
        try:
            user_input = float(input(f"{prompt} ({min_value}-{max_value}): "))
            if min_value <= user_input <= max_value:
                return user_input
            else:
                print(f"Input must be between {min_value} and {max_value}. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

def return_prediction(model, scaler, sample_dict):
    # Define the variables based on the dictionary in the desired order
    are = sample_dict['Area(total)']
    trm = sample_dict['Troom']
    bed = sample_dict['Nbedrooms']
    nws = sample_dict['Nbwashrooms']
    tws = sample_dict['Twashrooms']
    rar = sample_dict['Roof(Area)']
    law = sample_dict['Lawn(Area)']
    flr = sample_dict['Nfloors']
    api = sample_dict['API']
    anb = sample_dict['ANB']
    exp = sample_dict['EXPECTED']

    room = [[are, trm, bed, nws, tws, rar, law, flr, api, anb, exp]]
    room1 = scaler.transform(room)

    classes = np.array(['A:- 0', 'B:- 1', 'C:- 2', 'D:- 3', 'E:- 4'])
    class_ind = model.predict(room1)

    return classes[class_ind]

def ask_user():
    user_data = {}

    user_data['Area(total)'] = validate_input("Enter the total area", 290, 360)
    user_data['Troom'] = validate_input("Enter the number of rooms", 5, 9)
    user_data['Nbedrooms'] = validate_input("Enter the number of bedrooms", 2, 8)
    user_data['Nbwashrooms'] = validate_input("Enter the number of bedrooms with washrooms", 1, 7)
    user_data['Twashrooms'] = validate_input("Enter the number of washrooms", 3, 8)
    user_data['Roof(Area)'] = validate_input("Enter the roof area", 0, 120)
    user_data['Lawn(Area)'] = validate_input("Enter the lawn area", 70, 110)
    user_data['Nfloors'] = validate_input("Enter the number of floors", 1, 8)
    user_data['API'] = validate_input("Enter the API", 59, 99)
    user_data['ANB'] = validate_input("Enter the ANB", 1, 6)
    user_data['EXPECTED'] = validate_input("Enter the expected value", 0, 2041)

    return user_data

# Example usage
user_input = ask_user()
prediction = return_prediction(model=svc_pickle_model, scaler=person_scaler, sample_dict=user_input)
print(f"The predicted class is: {prediction}")
'''
```
